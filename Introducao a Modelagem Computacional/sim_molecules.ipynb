```python
import os

import numpy as np

import utils, odes
```

```python
OUTPUT_DIR = ".local/simulations/ABC Molecules"
CSV_FILE_PATH = os.path.join(OUTPUT_DIR, "simulation.csv")
PLOT_FILE_PATH = os.path.join(OUTPUT_DIR, "SCRIPT_HK1.png")
```

```python
os.makedirs(name=OUTPUT_DIR, exist_ok=True)
```

```python
time_steps, state_history = utils.simulate(
    ode_system=odes.ode_abc_molecules,
    initial_condition=np.array([ 50, 25, 0 ]),
    delta_t=0.1,
    t_final=1,
    solver=utils.rk4,
    ode_system_kwargs={ "k1": 0.3, "k2": 0.25 }
)
```

```python
utils.save_simulation(
    time_steps=time_steps,
    state_history=state_history,
    columns_labels=[ 'A', 'B', 'C' ],
    file_path=CSV_FILE_PATH
)
```

```python
utils.plot_simulation(
    time_steps=time_steps,
    state_history=state_history,
    xlabel="Tempo (Dias)",
    ylabel="Quantidade de Moleculas",
    title="Modelo de Formao de Moleculas",
    columns_to_plot=[ 0, 1, 2 ],
    columns_labels=[ 'A', 'B', 'C' ],
    figure_path=PLOT_FILE_PATH
)
```
