```python
import os

import numpy as np

import utils, odes
```

```python
OUTPUT_DIR = ".local/simulations/Competition"
CSV_FILE_PATH = os.path.join(OUTPUT_DIR, "simulation.csv")
PLOT_FILE_PATH = os.path.join(OUTPUT_DIR, "simulation.png")
```

```python
os.makedirs(name=OUTPUT_DIR, exist_ok=True)
```

```python
time_steps, state_history = utils.simulate(
    ode_system=odes.ode_competition,
    initial_condition=np.array([ 10, 10 ]),
    delta_t=0.01,
    t_final=200,
    solver=utils.rk4,
    ode_system_kwargs={ "r1": 0.2, "r2": 0.1, "w11": 0.02, "w12": 0.01, "w21": 0.02, "w22": 0.03 }
)
```

```python
utils.save_simulation(
    time_steps=time_steps,
    state_history=state_history,
    columns_labels=[ 'Sa', 'Sb' ],
    file_path=CSV_FILE_PATH
)
```

```python
utils.plot_simulation(
    time_steps=time_steps,
    state_history=state_history,
    xlabel="Tempo (Dias)",
    ylabel="Quantidade de Indivduos",
    title="Modelo de Competio entre Espcies",
    columns_to_plot=[ 0, 1 ],
    columns_labels=[ 'Sa', 'Sb' ],
    figure_path=PLOT_FILE_PATH
)
```
